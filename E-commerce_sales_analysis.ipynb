## E-commerce Sales Analysis

```python
'''
Problem Statement--

The company faces numerous challenges across its operations. To ensure sustainable growth and operational efficiency, it aims to address 
key areas:

Financial Performance: Enhance financial performance by analyzing total revenue, identifying top-selling products, calculating average order 
value, and identifying peak sales periods.

Customer Insights: Understand customer behavior by identifying top-performing locations, analyzing order cancellations, and 
differentiating between B2B and B2C customers.

Logistics and Fulfillment: Streamline logistics and fulfillment processes by calculating average shipping time, assessing variations in 
shipping time across service levels, and evaluating the percentage of orders fulfilled by Amazon vs Merchant.

Product Management: Enhance product management strategies by identifying popular product categories, analyzing average quantity 
ordered, and understanding the impact of promotions on sales.

Customer Satisfaction and Returns: Enhance customer satisfaction by analyzing return and cancellation rates, identifying problematic products
or categories, and addressing quality issues.

This comprehensive analysis aims to provide actionable insights, enabling data-driven decision-making, enhancing operational efficiency, 
and driving sustainable growth in the competitive retail landscape.
'''
```

```python
'''
Table Content--

1. Order ID: A unique identifier assigned to each order placed in the e-commerce system.
2. Date : Date on which product ordered.
3. Status : Current status of the order (e.g., Shipped, Cancelled, Delivered to Buyer, etc.)
4. Fulfillment: Describes how the order was fulfilled(e.g., Amazon, Merchant)
5. Sales Channel : The platform through which the sale was made(e.g., Amazon.in)
6. ship-service-level : Shipping service level (e.g., Standard, Expedited)
7. Style : Style of the product
8. SKU : Specific identifiers used for stock keeping and product design.
9. Category : Category of the product
10. Size : Size of the product
11. Quantity : The number of units of the product ordered
12. currency : The currency used for the transaction.
13. Amount : Amount of the transaction
14. ship-city, ship-state, ship-country: Details about the location to which the product was shipped.
15. ship-postal-code: Postal code where the product was shipped.
16. promotion_ids : Promotion IDs if any
17. B2B : A boolean indicating whether the order is Business-to-Business (B2B)
18. fulfilled_by : Who fulfilled the shipping (e.g., Easy Ship)
'''
```

## Data Cleaning

```python
# Creating copy for manipulation and processing
data = df.copy()
```

```python
data.head()
```

```python
# To check duplicates
data.duplicated().any()
```

```python
data.drop_duplicates(inplace = True)
```

```python
# Capitalizing these two columns to look better.
data["ship-city"] = data["ship-city"].str.capitalize()
data["ship-state"] = data["ship-state"].str.capitalize()
```

converting the column incorrect datetype :                                                                      
ship-postal-code to be in category but it is in float64.                            
converting B2B dtype from boolean to object.

```python
data["ship-postal-code"] = data["ship-postal-code"].astype('category')
data["B2B"] = data["B2B"].astype(str)

# replacing values with B2B (business-to-business)  and B2C (business-to-consumer)
# -> Amazon do E-commerce business in two ways:
# 1. by selling the product from seller to consumer(B2C)
# 2. by selling the product from seller to seller(B2B)

# In the B2B variable when it is False it treated as B2C and for True it is B2B.
```

```python
data["B2B"] = data["B2B"].replace({"False" : "B2C", "True" : "B2B"})
```

```python
# Renaming columns name in a appropriate manner
data.rename(columns = {"Order ID" : "Order_ID", "Date" : "Order_date", "Status" : "Order_status", "Fulfilment" : "Fulfillment", 
                       "Sales Channel ": "Sales_channel","ship-service-level" : "Shipping_type", "Qty" : "Quantity", 
                       "Courier Status":"Courier_Status","ship-city" : "Shipping_city", 
                       "ship-state" : "Shipping_state", "ship-postal-code" : "Pincode","ship-country": "Shipping_country",
                       "promotion-ids" : "Promotion_Ids", "B2B" : "Business_type" , "fulfilled-by" : "Fulfilled_by"}, inplace = True)
```

```python
# Replacing values of Order_status in more appropriate values, it will be easy to understand.
data['Order_status'] = data['Order_status'].replace({
    'Shipped - Delivered to Buyer': 'Delivered',
    'Shipped - Returned to Seller': 'Cancelled',
    'Shipped - Rejected by Buyer': 'Cancelled',
    'Shipped - Lost in Transit': 'Lost in Transit',
    'Shipped - Out for Delivery': 'Out for Delivery',
    'Shipped - Returning to Seller': 'Cancelled',
    'Shipped - Picked Up': 'Picked Up',
    'Pending - Waiting for Pick Up': 'Waiting for Pick Up',
    'Shipped - Damaged': 'Damaged'
})
```

Q6. Are there any trends in order cancellations?

```python
trends = data[['Category', 'Shipping_state', 'Size', 'Quantity' , 'Month_name']]
```

```python
# Cancelled Orders Analysis
cancelled_orders =data[data.Order_status == 'Cancelled']
for i,predictor in enumerate(trends.drop('Quantity',axis=1)):
    cancelled_reasons = cancelled_orders.groupby(predictor)['Quantity'].count().sort_values(ascending=False)
    plt.figure(figsize = (10,4))
    cancelled_reasons.plot.bar(color = 'salmon')
    plt.title("Reason for cancelled orders")
    plt.xlabel(predictor)
    plt.ylabel("No. of Cancelled Orders")
    plt.show()
```

The state Maharastra have more orders cancelled which is about 3500.                                                                 
The medium size dress were cancelled more.                                                                                       
In the month of April there are upto 8000 orders cancelled.

```python
# Calculate return rate

cancelled_orders = data[data['Order_status'] == 'Cancelled'].groupby('Month_name')['Order_status'].count()

return_rate = (cancelled_orders / total_orders * 100).fillna(0)  # Calculate return rate and replace NaN with 0

# Plot the return rate
plt.figure(figsize=(10, 4))
sns.lineplot(x=return_rate.index, y=return_rate.values, marker='o')
plt.grid(c = 'grey', ls = 'dashed', lw = 1)
plt.xticks(['March','April','May','June'])
plt.xlabel('Month')
plt.ylabel('Return Rate (%)')
plt.title('Monthly Return Rate')
plt.show()
```

Result : Return rate is more in the month of April i.e., 6.5 % which is reduced by 6.5% to 4.8% in June.

Q10. Are there specific products or categories that have a higher return rate?

```python
cancelled_orders = data[data['Order_status'] == 'Cancelled'].groupby('Category')['Order_status'].count()

return_rate = (cancelled_orders / total_orders * 100).fillna(0)  # Calculate return rate and replace NaN with 0

# Plot the return rate
plt.figure(figsize=(8, 6))
sns.lineplot(x=return_rate.index, y=return_rate.values, marker='o')
plt.grid(c = 'grey', ls = 'dashed', lw = 1)
plt.xlabel('Product')
plt.ylabel('Return Rate (%)')
plt.title('Product wise Return Rate')
plt.show() 
```

Result : The product 'Set' has more return rate of 6.5% and 'Kurta' has 6.3% of return rate.                                  
Sarees have low return rate which is close to 0 followed by Bottom,Blouse,Ethnic Dress.                                              
There is no return rate for Dupatta this may be due to the product is not returned and they are satisfied with the product.

## Conclusion

These are some of the quick insights from EDA analysis

1. Sales peaked in April but experienced a decline in May.
2. Sets, kurtas, and western dresses in size 'M' attracted the most customer interest, while only three orders were placed for Dupattas.
3. Bangalore saw the highest sales, closely followed by Hyderabad.
4. The majority of orders were placed on Sundays, indicating that customers might have more time to place order on Sunday.
5. Average order value for Business-to-Business (B2B) transactions exceeded that of Business-to-Customer (B2C) transactions, indicating higher order values from B2B clients.
6. Average shipping time was approximately one day, and Amazon fulfilled nearly 70% of orders.
7. Sets and kurtas had higher return and cancellation rates, aligning with customer purchase patterns.
8. Maharashtra had the highest cancellation rate among states.
